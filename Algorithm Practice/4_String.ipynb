# 替换空格

请实现一个函数，把字符串 s 中的每个空格替换成"%20"

链接：https://leetcode.cn/problems/ti-huan-kong-ge-lcof/

```python
class Solution:
    def replaceSpace(self, s: str) -> str:
        return s.replace(' ', '%20')
    
    def method_2(self, s: str) -> str: # 使用join + split
        return "%20".join(s.split(" "))
    
    def method_3(self, s: str) -> str: # 添加空列表，添加匹配的结果
        res = []
        for i in range(len(s)):
            if s[i] == ' ':
                res.append('%20')
            else:
                res.append(s[i])
        return ''.join(res)
    
if __name__ == '__main__':
    s = Solution()
    print(s.replaceSpace('We are happy.'))
    print(s.method_2('We are happy.'))
    print(s.method_3('We are happy.'))
```

# 左旋转字符串

字符串的左旋转操作是把字符串前面的若干个字符转移到字符串的尾部。请定义一个函数实现字符串左旋转操作的功能。

链接：https://leetcode.cn/problems/zuo-xuan-zhuan-zi-fu-chuan-lcof/

```python
class Solution:
    def reverseLeftWords(self, s: str, n: int) -> str:
        s = list(s)
        s[0:n] = list(reversed(s[0:n]))
        s[n:] = list(reversed(s[n:]))
        s.reverse()
        
        return "".join(s)

if __name__ == '__main__':
    s = Solution()
    print(s.reverseLeftWords("abcdefg", 2))
    print(s.reverseLeftWords("lrloseumgh", 6))
```

# KMP算法

- 参考链接
  - https://www.bilibili.com/video/BV1PD4y1o7nd/
  - https://www.bilibili.com/video/BV1M5411j7Xx/

**KMP的经典思想就是：当出现字符串不匹配时，可以记录一部分之前已经匹配的文本内容，利用这些信息避免从头再去做匹配。**

## 前缀表

- next数组就是一个前缀表（prefix table）
- 前缀是指不包含最后一个字符的所有以第一个字符开头的连续子串；后缀是指不包含第一个字符的所有以最后一个字符结尾的连续子串
- 前缀表是用来回退的，它记录了模式串与主串(文本串)不匹配的时候，模式串应该从哪里开始重新匹配，此也意味着在某个字符失配时，前缀表会告诉你下一步匹配中，模式串应该跳到哪个位置
- 例子：要在文本串 aabaabaafa 中查找是否出现过一个模式串 aabaaf
  - 文本串中第六个字符b 和 模式串的第六个字符f，不匹配了。如果暴力匹配，发现不匹配，此时就要从头匹配了
  - 但如果使用前缀表，就不会从头匹配，而是从上次已经匹配的内容开始匹配，找到了模式串中第三个字符b继续开始匹配
- **前缀表记录下标i之前（包括i）的字符串中，有多大长度的相同前缀后缀**

## 最长相等前后缀

- 对于模式串 aabaaf
  - 前缀包括：a，aa，aab，aaba，aabaa
  - 后缀包括：f，af，aaf，baaf，abaaf
- 最长相等前后缀，也就是常说的“最长公共前后缀”
- 字符串 a，既没有前缀，也没有后缀，最长相等前后缀为0
- 字符串 aa，前缀为 a，后缀为 a，最长相等前后缀为1
- 字符串 aab，前缀为 a，aa，后缀为 b，ab，最长相等前后缀为 0
- 字符串 aaba，前缀为 a，aa，aab，后缀为 a，ba，aba，最长相等前后缀为 1
- 字符串 aabaa，最长相等前后缀为 2（aa）
- 字符串 aabaaf， 最长相等前后缀为 0
- 所以对于模式串 aabaaf，得到的序列  `[0,1,0,1,2,0]` 就是前缀表

## 前缀表工作流程

对于文本串 aabaabaafa，模式串 aabaaf，得到前缀表 `[0,1,0,1,2,0]`

位置 f 发生冲突，不匹配，它前面的子串 aabaa 的最长相等前后缀为 2（aa），意味着子串 aabaa 有一个后缀 aa，前面也有一个与其相等的前缀 aa。此时在后缀的后面（f）发生冲突，那么就要找与其相等的前缀的后面继续开始匹配

那么就找到了下标2，指向 b，继续匹配

所以前缀表具有告诉我们当前位置匹配失败，跳到之前已经匹配过的地方的能力

模式串与前缀表对应位置的数字表示的就是：**下标i之前（包括i）的字符串中，有多大长度的相同前缀后缀**

## 前缀表与next数组

next数组就可以是前缀表，但是很多实现都是把前缀表统一减一（右移一位，初始位置为-1）之后作为next数组

那么对于文本串 aabaabaafa，模式串 aabaaf，得到前缀表 `[0,1,0,1,2,0]`，减一后变为 `[-1,0,-1,0,1,-1]`

这并不涉及到KMP的原理，而是具体实现，next数组既可以就是前缀表，也可以是前缀表统一减一（右移一位，初始位置为-1）

## 时间复杂度分析

其中 n 为文本串长度，m 为模式串长度，因为在匹配的过程中，根据前缀表不断调整匹配的位置，可以看出匹配的过程是 $O(n)$，之前还要单独生成next数组，时间复杂度是 $O(m)$。所以整个KMP算法的时间复杂度是 $O(n+m)$ 的

暴力的解法显而易见是 $O(n \times m)$，所以KMP在字符串匹配中极大地提高了搜索的效率

# 找出字符串中第一个匹配项的下标

给你两个字符串 haystack 和 needle ，请你在 haystack 字符串中找出 needle 字符串的第一个匹配项的下标（下标从 0 开始）。如果 needle 不是 haystack 的一部分，则返回  -1

示例
- 输入：haystack = "leetcode", needle = "leeto"
- 输出：-1
- 解释："leeto" 没有在 "leetcode" 中出现，所以返回 -1 


链接：https://leetcode.cn/problems/find-the-index-of-the-first-occurrence-in-a-string/

```python
from typing import List

class Solution:
    def strStr(self, haystack: str, needle: str) -> int: # 暴力法
        m, n = len(haystack), len(needle)
        for i in range(m):
            if haystack[i:i+n] == needle:
                return i
        return -1    
    
    def method_2(self, haystack: str, needle: str) -> int: # 使用 find
        return haystack.find(needle)
    
    def method_3(self, haystack: str, needle: str) -> int: # 使用 index
        try:   
            return haystack.index(needle)
        except ValueError:
            return -1
        
    def getNext_1(self, next: List[int], s: str) -> None: # 获取前缀表(不减一)
        j = 0 # 初始化j，用于指向前缀的当前比较位置
        next[0] = 0 # 第一个元素next[0]始终为0
        for i in range(1, len(s)): # 从第二个元素开始计算next数组
            while j > 0 and s[i] != s[j]: # 如果j > 0且两个字符不相等，那么j就要回溯
                j = next[j - 1]
            if s[i] == s[j]: # 如果当前字符匹配，将j向后移动一位
                j += 1
            next[i] = j # 将当前位置的j值记录在next数组中，表示当前位置的最长相同前后缀的长度
    
    def KMP_1(self, haystack: str, needle: str) -> int: # Knuth-Morris-Pratt (KMP) algorithm（前缀表不减一）
        if len(needle) == 0: # 如果needle字符串为空，直接返回0，因为任何字符串在空字符串中都出现过
            return 0
        next = [0] * len(needle) # 初始化next数组，用于记录在needle中每个位置的最长相同前后缀的长度（不减一）
        self.getNext_1(next, needle)  # 调用getNext函数模式串的前缀表
        j = 0 # 初始化j，用于指向needle字符串的当前比较位置
        for i in range(len(haystack)): # 遍历haystack字符串
            while j > 0 and haystack[i] != needle[j]: # KMP算法的关键部分，当当前字符不匹配时，根据next数组调整j的位置
                j = next[j - 1]
            if haystack[i] == needle[j]: # 如果当前字符匹配，将j向后移动一位
                j += 1
            if j == len(needle): # 如果j移动到needle字符串的末尾，说明找到了完全匹配的子串
                return i - len(needle) + 1 # 返回子串的起始坐标
        return -1 # 如果没有找到完全匹配的子串，返回-1表示找不到
    
    def getNext_2(self, next, s): # 获取前缀表(减一)
        j = -1
        next[0] = j
        for i in range(1, len(s)):
            while j >= 0 and s[i] != s[j+1]:
                j = next[j]
            if s[i] == s[j+1]:
                j += 1
            next[i] = j
    
    def KMP_2(self, haystack: str, needle: str) -> int: # Knuth-Morris-Pratt (KMP) algorithm（前缀表减一）
        if not needle:
            return 0
        next = [0] * len(needle)
        self.getNext_2(next, needle)
        j = -1
        for i in range(len(haystack)):
            while j >= 0 and haystack[i] != needle[j+1]:
                j = next[j]
            if haystack[i] == needle[j+1]:
                j += 1
            if j == len(needle) - 1:
                return i - len(needle) + 1
        return -1

if __name__ == '__main__':
    haystack = "hello"
    needle = "ll"
    solution = Solution()
    print(solution.strStr(haystack, needle))
    print(solution.method_2(haystack, needle))
    print(solution.method_3(haystack, needle))
    print(solution.KMP_1(haystack, needle))
    print(solution.KMP_2(haystack, needle))
```

# 重复的子字符串

给定一个非空的字符串 s ，检查是否可以通过由它的一个子串重复多次构成

链接：https://leetcode.cn/problems/repeated-substring-pattern/

```python
class Solution:
    def repeatedSubstringPattern(self, s: str) -> bool: # 暴力法
        n = len(s)
        if n <= 1:
            return False
        substr = ""
        for i in range(1, n//2 + 1):
            if n % i == 0:
                substr = s[:i]
                if substr * (n//i) == s:
                    return True
        return False
    
    def method_2(self, s: str) -> bool: # find
        n = len(s)
        if n <= 1:
            return False
        ss = s[1:] + s[:-1] 
        print(ss.find(s))              
        return ss.find(s) != -1
    
    # 当一个字符串由重复子串组成的，最长相等前后缀不包含的子串就是最小重复子串
    def getNext_1(self, nxt, s): # 前缀表 减一
        nxt[0] = -1
        j = -1
        for i in range(1, len(s)):
            while j >= 0 and s[i] != s[j+1]:
                j = nxt[j]
            if s[i] == s[j+1]:
                j += 1
            nxt[i] = j
        return nxt
     
    def KMP_1(self, s: str) -> bool: # KMP(前缀表减1)
        if len(s) == 0:
            return False
        nxt = [0] * len(s)
        self.getNext_1(nxt, s)
        if nxt[-1] != -1 and len(s) % (len(s) - (nxt[-1] + 1)) == 0:
            return True
        return False
    
    def getNext_2(self, nxt, s): #前缀表 不减一
        nxt[0] = 0
        j = 0
        for i in range(1, len(s)):
            while j > 0 and s[i] != s[j]:
                j = nxt[j - 1]
            if s[i] == s[j]:
                j += 1
            nxt[i] = j
        return nxt

    def KMP_2(self, s: str) -> bool: # KMP(前缀表不减1)
        if len(s) == 0:
            return False
        nxt = [0] * len(s)
        self.getNext_2(nxt, s)
        if nxt[-1] != 0 and len(s) % (len(s) - nxt[-1]) == 0:
            return True
        return False
    
if __name__ == "__main__":
    s = Solution()
    print(s.repeatedSubstringPattern("abab"))
    print(s.method_2("abab"))
    print(s.KMP_1("abab"))
    print(s.KMP_2("abab"))
```
